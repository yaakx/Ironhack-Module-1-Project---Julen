```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
df = pd.read_csv("../data/processed/forbes_clean.csv", index_col = 0)
```

```python
df
```

```python
r = requests.get("https://www.forbes.com/ajax/list/data?year=2018&uri=billionaires&type=person")
```

```python
df_2 = pd.DataFrame(r.json())
```

```python
df_2
```

```python
df_2 = df_2.loc[: , ["age", "name", "gender", "country"]]
```

```python
df_2
```

```python
df_def = df.drop(["age", "gender", "country"], axis=1)
```

```python
df_def = df_def.merge(df_2, on = ['name'], how = 'left')
```

```python
df_def.tail()
```

```python
df_def.to_csv("../data/processed/forbes_definitive.csv")
```
